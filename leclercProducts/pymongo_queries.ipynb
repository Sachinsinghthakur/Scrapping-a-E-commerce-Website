<h2>how many products have sale_price greater than 300

```python
query_pipeline_2=[   {
                       "$match":{
                          "sale_price": {
                                  "$gt": 300
                                       }
                                   }
                    },
                    {
                        "$count":"total"
                    }
                   ]
```

```python
total_documents=0
for collection in db.collection_names():
    for output in db[collection].aggregate(query_pipeline_2):
        output=dict(output)
        total_documents+=output["total"]
    
print(total_documents)
```
